```python
import pandas as pd
import numpy as np
import requests

url_franquicias = 'https://gist.githubusercontent.com/ahcamachod/68eede222bdf1ba0ca413d529e800e7c/raw/576b0f3856467817df0fcf041d4a0787579917b0/ventas_franquicia.json'

# Descargar el JSON desde la URL
response = requests.get(url_franquicias)
datos_json = response.json()  # Convertir la respuesta en un diccionario de Python

# Crear un DataFrame de Pandas
df = pd.DataFrame(datos_json)

# print(df)

# Mostrar las primeras filas
print(df.head())
df.head()
```

podemos solicitar nuevamente la ayuda de ChatGPT.

Prompt:

El DataFrame "df" tiene las columnas "ID_tienda", "ventas" y "tienda".

Las columnas "ventas" y "tienda" contienen diccionarios anidados dentro de cada una de las filas. ¿Cómo puedo transformar los datos de estos diccionarios y columnas para hacerlos más accesibles en mi DataFrame?

```python
df_ventas = pd.json_normalize(df['ventas'])
df_tienda = pd.json_normalize(df['tienda'])
# Concatenar los DataFrames con los datos originales (excluyendo las columnas anidadas)
df_final = pd.concat([df.drop(columns=['ventas', 'tienda']), df_ventas, df_tienda], axis=1)
# Mostrar las primeras filas del DataFrame final
print(df_final.head())
df_final.head()
```
